```python
import numpy as np
import pandas as pd
import requests 
from bs4 import BeautifulSoup as bs
import re
import nltk
import string
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

url='https://www.sciencedirect.com/science/article/abs/pii/S1361920999000309'
response = requests.get(url).text
soup = bs(response, "html.parser")

link = []

for a in soup.find_all('a', href=True):
    link.append(a['href'])
               
link.pop(0)
link
```

```python
from bs4 import BeautifulSoup as bs
data_text = []
for i in link:
    response = requests.get(i)
    soup = bs(response.content, features = 'lxml')
    text = soup.get_text()
    data_text.append(text)
data_text
```

```python
df1 = pd.DataFrame(data_text, columns = ['original Data'])
df1
```

```python
for j in range (len(data_text)):
    data_text[j]=re.sub(r'\W'," ",data_text[j])
    data_text[j]=re.sub(re.escape(string.punctuation)," ",data_text[j])
    data_text[j]=re.sub(r'\s+[a-zA-Z]\s+'," ",data_text[j])
    data_text[j]=re.sub(r'\^+[a-zA-Z]\s+'," ",data_text[j])
    data_text[j]=re.sub(r'\s+'," ",data_text[j])
    data_text[j].lower()
    data_text[j]  = np.array(data_text[j] )
d=data_text
d
```

```python
df2 = pd.DataFrame(d,columns = ['Cleaned Data'])
df2
```

```python
df=pd.DataFrame()
df['original data']=data_text
df['cleaned']=d
df

```

```python
df.to_csv('cleaned02.csv')
```

# Polarity and subjectivity

```python
df['Polarity']=df["cleaned"].apply(lambda x : TextBlob(x).sentiment[0])
df['Subjectivity']=df["cleaned"].apply(lambda x : TextBlob(x).sentiment[1])
```

```python
df['Length']=df['cleaned'].apply(lambda x: len(x.split()))
```

# Sentiment Analysis

```python
import nltk
import string
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.sentiment import SentimentIntensityAnalyzer
y=len(df.index)
neg=[]
pos=[]
neu=[]
com=[]
for i in range(y):
    def sa(i):
        score = SentimentIntensityAnalyzer().polarity_scores(i)
        v1=score['neg']
        positive= score['pos']
        neutral= score['neu']
        compound= score['compound']
        neg.append(v1)
        pos.append(positive)
        neu.append(neutral)
        com.append(compound)
    z=sa(df['cleaned'][i])
  
```

```python
df['Negative Sentiments']=neg
df['Postive Sentiments']=pos
df['Neutral Sentiments']=neu
df['compound Sentiments']=com
```

```python
df
```

```python
df.loc[df['Polarity'] >= 0.001, 'Sentiment p/n']=1
df.loc[df['Polarity'] <= -0.001, 'Sentiment p/n']=-1
df.loc[df['Polarity'] == 0.0, 'Sentiment p/n']=0
df['Sentiment p/n']
df
```
